# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
plt.show()
```

```python
y = [1, 2, 3, 4]
print(y)
```

```python
plt.plot(y)
plt.show()
```

```python
x = [1, 2, 3, 4]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.plot(x, y, 'or')
plt.show()
```

```python
# barchart
plt.bar(x, y)
plt.show()
```

```python
# piechart
plt.pie(x)
plt.title('This is a piechart')
plt.show()
```
